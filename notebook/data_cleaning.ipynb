```python
def clean_unicef_data(df):
    df = df.drop(columns=[
        'observation_confidentaility',
        'time_period_activity_related_to_when_the_data_are_collected',
        'unit_multiplier',
        'unit_of_measure'], errors='ignore') 
    
    df = df.rename(columns={
        'country': 'Country',
        'indicator': 'Indicator',
        'time_period': 'Year',
        'obs_value': 'Value',
        'sex': 'Gender'})
    
    df = df[df['Gender'] == 'Total'].reset_index(drop=True)
    return df
```

```python
data1_cleaned = clean_unicef_data(data1)
data2_cleaned = clean_unicef_data(data2)
```

```python
mno.bar(data1_cleaned, color='crimson')
```

```python
mno.bar(data2_cleaned, color='powderblue')
```

```python
data1_cleaned.to_csv('../data/cleaned/unicef_indicator_1_cleaned.csv', index=False)
data2_cleaned.to_csv('../data/cleaned/unicef_inidcator_2_cleaned.csv', index=False)
```
